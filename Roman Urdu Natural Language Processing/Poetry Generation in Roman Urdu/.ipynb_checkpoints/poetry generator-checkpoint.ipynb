### Building a corpus from the given 'poetry.txt' file

```python
from google.colab import files

def read_file(filename):
  uploaded = files.upload()

  a_file = open(filename, "r")

  str_final = ""

  for line in a_file:
    str_stripped = line.rstrip()
    str_final += ' ' + str_stripped
  
  a_file.close()

  return str_final
```

```python
corpus=read_file('poetry.txt')
```

```python
corpus
```

### Creating a spaCy urdu language object

```python
urdu=spacy.blank('ur')
urdu
```

```python
urdu_text=urdu(corpus)
urdu_text
```

### Creating bigrams and trigrams from the corpus using a single n_grams function

```python
def n_grams(token, n):
  return list((token[i:i+n] for i in range(len(token)-n+1)))
```

```python
bigrams=n_grams(urdu_text, 2)
print(bigrams)
```

```python
trigrams=n_grams(urdu_text, 3)
print(trigrams)
```

### Conditonal Frequency Distribution

```python
from nltk.probability import ConditionalFreqDist

con_fre_dist = ConditionalFreqDist()

for w in bigrams:
  condition = len(w)
  con_fre_dist[condition][w] += 1 

# con_fre_dist
```

### Generating a single line of verse consisting of 6 to 8 words in a single line

```python
def generate_verse():
  verse_size = random.randint(6,8)
  verse = ''

  for i in range(verse_size):
      word = random.choice(urdu_text)

      if word in con_fre_dist[2].keys():
          word = random.choice(con_fre_dist[word].keys())

      if (word.text != ' ') or (word.text != '۔') or (word.text != ',') or (word.text != '"'): 
        verse += word.text + ' '
        
  return verse
```

### Creating a stanza containing 2 verses

```python
def generate_stanza(): 
  stan=""

  for i in range(2):
  # return generate_verse()
    stan += generate_verse()
    stan += '\n'
  return stan
```

```python
print(generate_stanza())
```

### Generating 7 stanzas, each stanza consisting of 2 verses

```python
for i in range(7):
  print(generate_stanza())
  print(' ')
```

### urdu2roman() function from Assignment 1

```python
roman_unicode_list = {
   u"\u0627":"A",
            u"\u0627":"A", 
            u"\u0675":"A", 
            u"\u0673":"A", 
            u"\u0630":"A", 
            u"\u0622":"AA", 
            u"\u0628":"B", 
            u"\u067E":"P", 
            u"\u062A":"T", 
            u"\u0637":"T", 
            u"\u0679":"T", 
            u"\u062C":"J", 
            u"\u0633":"S", 
            u"\u062B":"S", 
            u"\u0635":"S", 
            u"\u0686":"CH", 
            u"\u062D":"H", 
            u"\u0647":"H", 
            u"\u0629":"H", 
            u"\u06DF":"H", 
            u"\u062E":"KH", 
            u"\u062F":"D", 
            u"\u0688":"D", 
            u"\u0630":"Z", 
            u"\u0632":"Z", 
            u"\u0636":"Z", 
            u"\u0638":"Z", 
            u"\u068E":"Z", 
            u"\u0631":"R", 
            u"\u0691":"R", 
            u"\u0634":"SH", 
            u"\u063A":"GH", 
            u"\u0641":"F", 
            u"\u06A9":"K", 
            u"\u0642":"K", 
            u"\u06AF":"G", 
            u"\u0644":"L", 
            u"\u0645":"M", 
            u"\u0646":"N", 
            u"\u06BA":"N", 
            u"\u0648":"O", 
            u"\u0649":"Y", 
            u"\u0626":"Y", 
            u"\u06CC":"Y", 
            u"\u06D2":"E", 
            u"\u06C1":"H",
            u"\u064A":"E"  ,
            u"\u06C2":"AH"  ,
            u"\u06BE":"H"  ,
            u"\u0639":"A"  ,
            u"\u0643":"K" ,
            u"\u0621":"A",
            u"\u0624":"O",
            u"\u060C":""
}
```

```python
def urdu2roman(sen):
  if (detect(sen) != 'ur'):
    return
  
  sen.encode().decode('utf-8')

  for k,v in roman_unicode_list.items():
    sen=sen.replace(k,v)
    # print(k)
    # print(v)
  
  return sen
```

## Final output. 

```python
for i in range(7):
  print(urdu2roman(generate_stanza()))
  # print(generate_stanza())
  print(' ')
```
