# Python Crawler to scrap Google Images

### Beautiful Soup

```python
import requests
```

```python
def search_google_images(query):
    url = f"https://www.google.com/search?q={query}&tbm=isch"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    response = requests.get(url, headers=headers)
    return response.text
```

```python
from bs4 import BeautifulSoup

def extract_image_urls(html):
    soup = BeautifulSoup(html, "html.parser")
    image_urls = []
    for img in soup.find_all("img"):
        image_urls.append(img.get("src"))
    return image_urls
```

```python
if __name__ == "__main__":
    search_query = input("Enter the search query: ")
    search_results = search_google_images(search_query)
    image_urls = extract_image_urls(search_results)
    for i, url in enumerate(image_urls, start=1):
        print(f"{i}. {url}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

def get_image_urls(query):
    url = f"https://www.google.com/url?sa=i&url=https%3A%2F%2Fencrypted-tbn0.gstatic.com%2Flicensed-image%3Fq%3Dtbn%3AANd9GcTSFu1Jp6Sr1xplTnlX6Pi4qCUBOLga2fDSxg7T7XpOUADT14l7ZzVWljzaBZbu4CpAUvE-OhzWDkj-HGE&psig=AOvVaw0x2MJTXw22_Ol0eyWzs985&ust=1698766627609000&source=images&cd=vfe&ved=0CBEQjRxqFwoTCJCU8IKNnoIDFQAAAAAdAAAAABAE"

    driver = webdriver.Chrome() 
    driver.get(url)

    driver.implicitly_wait(10)  

    image_elements = driver.find_elements(By.TAG_NAME, 'img')

    image_urls = [img.get_attribute('src') for img in image_elements]

    driver.quit()

    return image_urls
```
